```python
import pandas as pd
```

```python
weather=pd.read_csv('Weather Training Data.csv')
weather.head(15)
```

```python
weather.info()
```

```python
weather.isnull().mean()*100
```

```python
# Checking which features are in non-numeric 
for label, content in weather.items():
    if pd.api.types.is_object_dtype(content):
        print(label)
```

```python
# Some of our data is in string datatype. We need to convert it to numbers.
# for label, content in weather.items():
#     if pd.api.types.is_object_dtype(content):
#         weather[label] = content.astype("category").cat.as_ordered()
```

```python
weather.dtypes
```

```python
#fill missing continous features
continous_columns = weather.select_dtypes(include=['float64','int64']).columns
for c in continous_columns:
    mean = weather[c].mean()
    weather[c]=weather[c].fillna(mean)
```

```python
#fill missing categorical features
categorical_columns = weather.select_dtypes(include=['object']).columns
for c in categorical_columns:
    most_frequent_value = weather[c].mode()[0]
    weather[c]=weather[c].fillna(most_frequent_value)
```

```python
#Now we have got no missing data in our columns and all of our data is numeric.
weather.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for c in categorical_columns:
    label_encoders[c] = LabelEncoder()
    weather[c]=label_encoders[c].fit_transform(weather[c])

weather.head()
```

```python
weather.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Let's visualize it more
corr_matrix = weather.corr()
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.heatmap(corr_matrix,
                 annot = True,
                 linewidth = 0.5,
                 fmt = ".2f",
                 cmap = sns.cubehelix_palette(rot=-.4))
```

```python
weather.drop(['row ID','Pressure9am', 'Temp3pm'], axis = 1, inplace = True)
```

```python
weather.head(15).T
```

```python
#split features and target
X=weather.drop(['RainTomorrow'], axis=1)
y=weather['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rforest_model = RandomForestClassifier()
rforest_model.fit(X_train, y_train)
```

```python
rforest_model.predict(X_val)
```

```python
rforest_model_score = rforest_model.score(X_val, y_val)
rforest_model_accuracy = round(rforest_model_score*100, 2)
print("The classification accuracy of Random Forest model is "+ str(rforest_model_accuracy)+"%")
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np
y_pred = rforest_model.predict(X_val)
cm = confusion_matrix(y_val,y_pred)
axes = sns.heatmap(cm, annot=True, annot_kws={"size": 16}, fmt='g', linewidths=.5)
class_labels = ['Not Rain', 'Rain']

axes.set_xlabel('Predicted', fontsize=12)
axes.set_ylabel('Actual', fontsize=12)

xtick_marks = np.arange(len(class_labels)) + 0.5
ytick_marks = np.arange(len(class_labels)) + 0.5

axes.set_xticks(xtick_marks)
axes.set_xticklabels(class_labels, rotation=45)

axes.set_yticks(ytick_marks)
axes.set_yticklabels(class_labels, rotation=0)

axes.set_title('Random Forest Confusion Matrix', fontsize=14, pad=20)
```

```python
#Hyperparameter Optimization of Random Forest Classifier
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [100, 150, 200]
max_depth = [10, 15, 20]
criterion = ['gini', 'entropy']
bootstrap = [True,False]
random_state = [10, 20]
max_features = ['auto', 'sqrt']
min_samples_split = [1, 2, 3]
min_samples_leaf = [1, 2, 3]

hyperF = dict(n_estimators = n_estimators, max_depth = max_depth,
              criterion = criterion, bootstrap = bootstrap,
              random_state = random_state, max_features = max_features, 
              min_samples_split = min_samples_split,
              min_samples_leaf = min_samples_leaf)

randF = RandomizedSearchCV(rforest_model, hyperF, cv = 3, verbose = 1, n_jobs = -1)
bestF = randF.fit(X_train, y_train)
```

```python
print("Results from Random Search ::" )
print("\nThe best estimator across ALL searched params:\n", randF.best_estimator_)
print("\nThe best parameters across ALL searched params:\n", randF.best_params_)
```

```python
rforest_model = RandomForestClassifier(**randF.best_params_)
rforest_model.fit(X_train, y_train)
```

```python
rforest_model.predict(X_val)
```

```python
rforest_model_score = rforest_model.score(X_val, y_val)
rforest_model_accuracy = round(rforest_model_score*100, 2)
print("The classification accuracy of Random Forest model is "+ str(rforest_model_accuracy)+"%")
```
